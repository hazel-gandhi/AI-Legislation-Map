I have exported both my datasets: one with the names of the bill and details of sponsor and the other with the names of the companies that have lobbied for the bill as csvs. I am now importing them here and merging them. 
I will then do some cleaning up and bring in the google maps API to gets lats and longs of states

```python
import pandas as pd
df1 = pd.read_csv("issues_and_clients_final - Sheet1.csv")
df1
```

```python
df2 = pd.read_csv("congress-ai-bills_final - Sheet1.csv")
df2.rename(columns={'desc': 'Issue'}, inplace=True)
df2
```

```python
merged_df = pd.merge(df1, df2, on='Issue', how='right')

df = merged_df
df
```

```python
#extracting state names from the sponsor's name to get lats and longs
```

```python
import re
```

```python
#cleaning up sponsor names to remove district numbers after their states to make the data even
```

```python
def clean_sponsor(sponsor):
    return re.sub(r'-\d+\]', ']', sponsor)
```

```python
df['sponsor'] = df['sponsor'].apply(clean_sponsor)
df
```

```python
#defining state names to convert abbreviates to full state names
```

```python
state_names = {
    'AL': 'Alabama', 'AK': 'Alaska', 'AZ': 'Arizona', 'AR': 'Arkansas',
    'CA': 'California', 'CO': 'Colorado', 'CT': 'Connecticut', 'DE': 'Delaware',
    'FL': 'Florida', 'GA': 'Georgia', 'HI': 'Hawaii', 'ID': 'Idaho',
    'IL': 'Illinois', 'IN': 'Indiana', 'IA': 'Iowa', 'KS': 'Kansas',
    'KY': 'Kentucky', 'LA': 'Louisiana', 'ME': 'Maine', 'MD': 'Maryland',
    'MA': 'Massachusetts', 'MI': 'Michigan', 'MN': 'Minnesota', 'MS': 'Mississippi',
    'MO': 'Missouri', 'MT': 'Montana', 'NE': 'Nebraska', 'NV': 'Nevada',
    'NH': 'New Hampshire', 'NJ': 'New Jersey', 'NM': 'New Mexico', 'NY': 'New York',
    'NC': 'North Carolina', 'ND': 'North Dakota', 'OH': 'Ohio', 'OK': 'Oklahoma',
    'OR': 'Oregon', 'PA': 'Pennsylvania', 'RI': 'Rhode Island', 'SC': 'South Carolina',
    'SD': 'South Dakota', 'TN': 'Tennessee', 'TX': 'Texas', 'UT': 'Utah',
    'VT': 'Vermont', 'VA': 'Virginia', 'WA': 'Washington', 'WV': 'West Virginia',
    'WI': 'Wisconsin', 'WY': 'Wyoming', 'DC': 'District of Columbia'
}
```

```python
#defining function to extract full state names from the sponsor name
```

```python
def extract_full_state(sponsor):
    match = re.search(r'\[.*-(\w{2})(?:-\d+)?\]', sponsor)
    if match:
        state_abbr = match.group(1)  
        return state_names.get(state_abbr, 'Unknown State')  
    return 'Unknown State'
```

```python
#making party name list to use in function below while converting abbreviations to full party name from sponsor column
```

```python
party_names = {
    'D': 'Democratic Party',
    'R': 'Republican Party',
    'IN': 'Independent',
    'L': 'Libertarian Party',
    'G': 'Green Party',
    'C': 'Constitution Party',
    'W': 'Working Families Party',
    'U': 'Unknown Party',  
}
```

```python
#defining function to extract full party name from sponsor column
```

```python
def extract_party(sponsor):
    match = re.search(r'\[.*-(\w{1,3})-', sponsor)
    if match:
        party_abbr = match.group(1)
        return party_names.get(party_abbr, 'Unknown Party') 
    return 'Unknown Party'
```

```python
#finally making new dataframe with separate columns for state and party
```

```python
df['state'] = df['sponsor'].apply(extract_full_state)
df['party'] = df['sponsor'].apply(extract_party)
df
```

```python
#more clean-up, removing party name and state from sponsor column since we already have that info in separate columns
```

```python
df['sponsor'] = df['sponsor'].str.replace(r'\[.*?\]', '', regex=True).str.strip()
df
```

```python
#just renaming some columns
```

```python
df.rename(columns={"Issue": "bill_name", "Clients": "lobbying_companies", "title": "bill_no"}, inplace=True)
df
```

```python
df['no_bills_in_state'] = df.groupby('state')['bill_name'].transform('size')
df
```

```python
def assign_color(state_bills):
    if state_bills < 3:
        return "#00CCE7"
    if state_bills < 10:
        return "#00AEC5"
    if state_bills < 20:
        return "#0090A3"
    else:
        return "#007281"
```

```python
df['properties.color'] = df['no_bills_in_state'].apply(lambda x: assign_color(x))
df
```

```python
df['lobbying_companies'] = df['lobbying_companies'].fillna("No Companies Available")
df
```
